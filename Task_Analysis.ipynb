Parsing the data to makdown fromat using llamaparse with instruction.

```python
from llama_parse import LlamaParse

instruction = """The document contains tabular data representing a company's business plan.
It includes detailed financial information such as values, formulas, expense types, dates, and other relevant textual details.
"""

parser = LlamaParse(
    api_key="",
    result_type="markdown",
    parsing_instruction=instruction,
    max_timeout=5000,
)

llama_parse_documents = await parser.aload_data("/content/Company 1 Business Plan (1).xlsx")
```

```python
parsed_doc = llama_parse_documents[0]
```

```python
# display the file to see how it look like

from IPython.display import Markdown

Markdown(parsed_doc.text)
```

Save the parsed data to a file

```python
from pathlib import Path

document_path = Path("/content/data_parsed.md")
with document_path.open("a") as f:
    f.write(parsed_doc.text)
```

```python
#load the saved document

from langchain_community.document_loaders import UnstructuredMarkdownLoader

loader = UnstructuredMarkdownLoader(document_path)
loaded_documents = loader.load()
```

Split the data into chunks. I recommend skipping this step initially when running the notebook to observe if there is a noticeable difference in the output. I included this step because my API key had usage limits.

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1048, chunk_overlap=200)
docs = text_splitter.split_documents(loaded_documents)
len(docs)
```

Hugging Face Embeddings

```python
from langchain.embeddings import HuggingFaceEmbeddings

# Initialize instructor embeddings using the Hugging Face model
embeddings = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
```

Vector store using FAISS

```python
from langchain.vectorstores import FAISS

# Create a FAISS instance for vector database from 'data'
vectordb = FAISS.from_documents(documents=docs,
                                 embedding=embeddings)

# Create a retriever for querying the vector database
retriever = vectordb.as_retriever(score_threshold = 0.7)
```

Created RetrievalQA chain along with prompt template

```python
from langchain.prompts import PromptTemplate
from langchain.chains import RetrievalQA
from langchain.llms import OpenAI


llm = OpenAI(api_key="", temperature=0.7)

prompt_template = """Given the following context and a question, generate an answer based on this context only.
In the answer try to provide as much text as possible from "response" section in the source document context without making much changes.
If the answer is not found in the context, kindly state "I don't know." Don't try to make up an answer.

CONTEXT: {context}

QUESTION: {question}"""


PROMPT = PromptTemplate(
    template=prompt_template, input_variables=["context", "question"]
)
chain_type_kwargs = {"prompt": PROMPT}


from langchain.chains import RetrievalQA

chain = RetrievalQA.from_chain_type(llm=llm,
                            chain_type="stuff",
                            retriever=retriever,
                            input_key="query",
                            return_source_documents=False,
                            chain_type_kwargs=chain_type_kwargs)
```

asking some questions

```python
chain('Return the total Value of Production for the year 2026')
```

```python
chain ("Return the methodology for calculating the the 'operating income' item for the year 2024")
```

The answers generated by the llm above is wrong, this shows that the llm was hallucinating.

Lets try to use few shot learning to fix the issue.

```python
few_shots = [
    {
        'Question': "Return the total Value of Production for the year 2026.",
        'SpreadsheetQuery': "SELECT Value_of_Production FROM CE_sheet WHERE year = 2026",
        'SpreadsheetResult': "Result of the query",
        'Answer': "75,260"
    },

    {
        'Question': "Check if the company has financing with a Covid moratorium and report them in a table.",
        'SpreadsheetQuery': "SELECT financing_details FROM BP_sheet WHERE financing_type LIKE '%Covid moratorium%'",
        'SpreadsheetResult': "Result of the query",
        'Answer': "Table with n rows for each financing with covid moratorium and m columns to provide all the information regarding the financing"
    },


    {
        'Question': "Return the methodology for calculating the 'services' item for the year 2025.",
        'SpreadsheetQuery': "SELECT methodology FROM calculation_methods_sheet WHERE item = 'services' AND year = 2025",
        'SpreadsheetResult': "Result of the query",
        'Answer': "The cost for services is the sum of the sub-items x,y,z. Sub-item x is calculated by multiplying the revenues of the same year by the average of the historical incidences of sub-item x on revenues. Sub-item y is calculated by taking the value of the previous year and multiplying it by the Italian inflation rate i%."
    },

    {
        'Question': "Return the methodology for calculating the the 'operating income' item for the year 2024",
        'SpreadsheetQuery': "SELECT methodology FROM calculation_methods_sheet WHERE item = 'operating income' AND year = 2024",
        'SpreadsheetResult': "Result of the query",
        'Answer': "The revenue for 2024 is calculated as the sum of two figures: the first are the orders already in existence of the company and are taken from the company's backlog; the second is calculated as a multiplication between last year's revenues and a growth rate defined by the company, from which the revenues from the backlog for 2024 are subtracted"
    }
]
```

Creating Semantic Similarity Based example selector


- create embedding on the few_shots

- Store the embeddings in Chroma DB

- Retrieve the the top most Semantically close example from the vector store

```python
from langchain.prompts import SemanticSimilarityExampleSelector
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma
embeddings = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
to_vectorize = [" ".join(example.values()) for example in few_shots]
to_vectorize
```

```python
vectorstore = Chroma.from_texts(to_vectorize, embeddings, metadatas=few_shots)
```

```python
example_selector = SemanticSimilarityExampleSelector(
    vectorstore=vectorstore,
    k=2,
)

example_selector.select_examples({"Question": "Return the total Value of Production for the year 2026."})
```

```python
from langchain.prompts import FewShotPromptTemplate
from langchain.prompts.prompt import PromptTemplate

example_prompt = PromptTemplate(
    input_variables=["Question", "SpreadsheetQuery", "SpreadsheetResult","Answer",],
    template="\nQuestion: {Question}\nSpreadsheetQuery: {SpreadsheetQuery}\nSpreadsheetResult: {SpreadsheetResult}\nAnswer: {Answer}",
)
```

```python
few_shot_prompt = FewShotPromptTemplate(
    example_selector=example_selector,
    example_prompt=example_prompt,
    suffix="Question: {input}",
    input_variables=["input"]
)
```

asking some questions

```python
print(llm(few_shot_prompt.format(input="Return the total Value of Production for the year 2026")))
```

```python
print(llm(few_shot_prompt.format(input="Return the total Value of Production for the year 2025")))
```
